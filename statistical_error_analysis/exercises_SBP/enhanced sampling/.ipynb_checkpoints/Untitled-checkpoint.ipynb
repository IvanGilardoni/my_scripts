## Thermodynamic integration
Write a MD or Langevin code sampling the canonical probability distribution at
a temperature T=1 of a three-dimensional potential of the form

V(x,y,z)=-5 log( 3 exp(-2x^2-1/4 y^2 -z^2) + 2 exp (-(x-3/2)^2-(y-5/2)^2-z^2))

```python
import numpy as np
import matplotlib.pyplot as plt
from numba import njit
from scipy.optimize import curve_fit
```

```python
def gradient(x,y,z):
    exp1=np.exp(-2*x**2-1/4*y**2-z**2)
    exp2=np.exp(-(x-3/2)**2-(y-5/2)**2-z**2)
    gradx=5*(12*x*exp1+4*(x-3/2)*exp2)/(3*exp1+2*exp2)
    grady=5*(3/2*y*exp1+4*(y-5/2)*exp2)/(3*exp1+2*exp2)
    gradz=5*(6*z*exp1+4*z*exp2)/(3*exp1+2*exp2)
    return [gradx,grady,gradz]
```

```python
njit(fastmath=True) ## why it gives me error?
def run(D,dt,N,q=[0,0,0],kT=1):#,seed=0):
    #seed=0
    #np.random.seed(seed)
    traj=[]
    traj.append(q)
    
    for i in range(N):
        grad=gradient(q[0],q[1],q[2])
        rando=[np.random.normal(),np.random.normal(),np.random.normal()]
        q=q-np.multiply(grad,D/kT*dt)+np.multiply(rando,np.sqrt(2*D*dt))
        ##[q[0]-D*grad[0]*dt+np.sqrt(2*D*dt)*np.random.normal(),q[1]-D*grad[1]*dt+np.sqrt(2*D*dt)*np.random.normal(),q[2]-D*grad[2]*dt+np.sqrt(2*D*dt)*np.random.normal()
        traj.append(q)
        # traj.append((i*dt),q)
    return np.array(traj)
```

```python
dt=0.001
Tmax=20
q0=[1.5,2.5,0]
N=round(Tmax/dt)
D=1
kT=1

time=np.linspace(0,Tmax,N+1)
traj=run(D,dt,N,q0,kT)
plt.figure(figsize=(10,6))
plt.plot(time,traj[:,0],label='x')
plt.plot(time,traj[:,1],label='y')
plt.plot(time,traj[:,2],label='z')
plt.legend()
```

```python
def gradient2(r,theta,z):
    cos=np.cos(theta)
    sin=np.sin(theta)
    exp1=np.exp(-2*r**2*cos**2-1/4*r**2*sin**2-z**2)
    exp2=np.exp(-(r*cos-3/2)**2-(r*sin-5/2)**2-z**2)
    gradz=5*(6*z*exp1+4*z*exp2)/(3*exp1+2*exp2)
    gradtheta=-5*(45/4*(r**2)*cos*sin*exp1+2*r*(-3*sin+5*cos)*exp2)/(3*exp1+2*exp2)
    Q=-5/(4*r**2)*(3*(8*r**2*cos**2+r**2*sin**2)*exp1+8*r*(cos*(r*cos-3/2)+sin*(r*sin-5/2))*exp2)/(3*exp1+2*exp2)
    return [gradtheta,gradz,Q]
```
