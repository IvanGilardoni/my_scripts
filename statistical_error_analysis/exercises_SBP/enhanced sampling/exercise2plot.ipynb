## Plot exp(-F), exercise 2

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
import seaborn as sns
```

```python
x=np.arange(0.01,5,0.01)
y=np.exp(-x)-2*np.log(x)

plt.figure(figsize=(10,6))
plt.plot(x,y)
plt.xlabel('s')
plt.ylabel('F(s)')
plt.grid()

plt.figure(figsize=(10,6))
plt.plot(x,np.exp(-y))
plt.xlabel('s')
plt.ylabel('exp(-F(s))')
plt.grid()
```

```python
x=np.arange(-1,2.5,0.1)
y=np.arange(-1.5,4,0.1)
x,y=np.meshgrid(x,y)
z=6*np.exp(-2*x**2-1/4*y**2)+4*np.exp(-(x-3/2)**2-(y-5/2)**2)

plt.figure(figsize=(10, 8))
#plt.subplot(1, 2, 1)
cs = plt.contour(x, y, z)
plt.clabel(cs)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
#plt.subplot(1, 2, 2)
plt.figure(figsize=(10,8))
cs = plt.contourf(x, y, z, cmap='coolwarm')
plt.colorbar(cs)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
zx=np.sum(z,axis=0)
zy=np.sum(z,axis=1)

plt.plot(zx)
plt.plot(zy)

plt.figure()
plt.plot(-np.log(zx))
plt.plot(-np.log(zy))
```
