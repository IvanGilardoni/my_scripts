## Visualization of the trajectory

Notebook to visualize the trajectory from a simulation with `simplemd`

For visualizing the trajectory we will use `py3Dmol` and the method `write_trajectory` of `simplemd`.

1. after running the simulation with `simplemd`, generate a text file with the trajectory using the method `write_trajectory` of `simplemd` and save it to your working directory as a `.xyz` file.

`simplemd.write_trajectory(path + "name_to_save.xyz", runner_variable.trajectory)`

2. visualize the trajectory using `py3Dmol` with the following steps:

- open the saved trajectory and process it as a text file

`with open(path+"/name_saved_trajectory.xyz") as f:
    traj_xyz = f.read()`

- visualize your text file trajectory using `py3Dmol`, here is a detailed description of its functioning https://william-dawson.github.io/using-py3dmol.html, https://pypi.org/project/py3Dmol/

`view = py3Dmol.view(width=400, height=300)`

this creates the object that contains the visualization window (in this case a box 400 x 300 pixels)

`view.addModelsAsFrames(traj_xyz, "xyz")`

this adds the frames contained in the trajectory to the visualization box

`view.setStyle({"sphere": {"radius":0.4}})`

this will set the particles to be visualized as spheres

`view.animate({'loop': "forward"})`

this will create a looping video out of your trajectory

`view.zoomTo()`
`view.show()`

these last two lines will display the movie.


```python
with open('output2.xyz') as f:
    traj_xyz = f.read()
```

```python
import py3Dmol
```

```python
view = py3Dmol.view(width=400, height=300)
```

```python
view.addModelsAsFrames(traj_xyz, "xyz")

view.setStyle({"sphere": {"radius":0.4}})
view.animate({'loop': "forward"})

view.zoomTo()
view.show()
```
