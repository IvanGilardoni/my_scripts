## Jarzynski Equality and Monte Carlo

```python
import matplotlib.pyplot as plt
import numpy as np
import random as rm
```

```python
def MC(kT=1,nsteps=1000,deltat=0.01,q0=0.0,k=1,c=1):
    q=q0  
    ene=0.5*k*(q-c)**2
    traj=[]
    for istep in range(nsteps):
        qtry=q+(2*np.random.rand()-1)*deltat
        enetry=0.5*k*(qtry-c)**2
        acc=np.exp(-(enetry-ene)/kT)
        if acc>1.0:
            acc=1.0
        if acc>np.random.rand():
            q=qtry
            ene=enetry
        traj.append(q)
    return np.array(traj)
```

```python
traj=MC()
plt.plot(traj)
```

```python
plt.figure(2)
nbins=100
n, bins, patches = plt.hist(traj, nbins,facecolor='green', alpha=0.5)
print(patches)
plt.grid(axis='both')
plt.show()
```

```python
def energy(k,q,c):
    return 0.5*k*(q-c)**2
```

```python
def MC2(kT=1,nsteps=1000,delta=10,q0=0.0,k=1,q1=10.0):#,c=0,kf=50):#,k=1,q1=10.0):
    q=q0
    icount=0
    #c=icount/nsteps*q1
    cold=0
    ## ene0=0.5*k*q**2
    work=0
    
    traj=[]
    #cs=[]
    #ks=[]
    
    for istep in range(nsteps):
        ###
        icount+=1
        #k=1+icount/nsteps*(kf-1)
        c=icount/nsteps*q1
        ene=0.5*k*(q-c)**2
        ### work+=ene-ene0
        ###
        
        qtry=q+(2*np.random.rand()-1)*delta
        
        enetry=0.5*k*(qtry-c)**2
        acc=np.exp(-(enetry-ene)/kT)
        if acc>1.0:
            acc=1.0
        if acc>np.random.rand():
            q=qtry
            ene=enetry
        #ene0=ene
        work+=ene-0.5*k*(q-cold)**2
        cold=c
        
        traj.append(q)
        
        #cs.append(c)
        #ks.append(k)
        
    return np.array(traj),work#,np.array(ks),work
```

```python
def MC3(kT=1,nsteps=1000,delta=1,q0=0.0,k=1,q1=10.0,c0=0,bw=False):#,c=0,kf=50):#,k=1,q1=10.0):
    q=q0
    icount=0
    #c=icount/nsteps*q1
    if bw:
        kold=2
    else:
        kold=1
    #kold=1
    ks=[]
    ## ene0=0.5*k*q**2
    work=0
    
    traj=[]
    #cs=[]
    #ks=[]
    
    for istep in range(nsteps):
        ###
        icount+=1
        if bw:
            k=2-icount/nsteps
        else:
            k=1+icount/nsteps
        #c=icount/nsteps*q1
        ene=0.5*k*(q-c0)**2
        ### work+=ene-ene0
        ###
        
        qtry=q+(2*np.random.rand()-1)*delta
        
        enetry=0.5*k*(qtry-c0)**2
        acc=np.exp(-(enetry-ene)/kT)
        if acc>1.0:
            acc=1.0
        if acc>np.random.rand():
            q=qtry
            ene=enetry
        #ene0=ene
        work+=ene-0.5*kold*(q-c0)**2
        kold=k
        
        traj.append(q)
        
        #cs.append(c)
        ks.append(k)
        
    return np.array(traj),work#,np.array(ks),work
```

```python
traj,work=MC2()
#print(cs)
plt.plot(traj)
print(work)
```

```python
#avworks=np.zeros(len(Ns))
#avexpworks=np.zeros(len(Ns))
works=[]
for i in range(1000):
    traj,work=MC2(delta=1,nsteps=1000)
    works.append(work)
        ### expworks[j]+=np.exp(-work/kT)
    ## print(works)
avworks=np.mean(works)
works=np.array(works)
avexpworks=np.mean(np.exp(-works))
print(avworks)
print(avexpworks)
```

## change c

```python
Ns=[1,10,20,50,100,200,500,1000,2000,5000,10000]

avworks=np.zeros(len(Ns))
avexpworks=np.zeros(len(Ns))

for j in range(len(Ns)):
    works=[]
    for i in range(Ns[j]):
        traj,work=MC2(delta=1,nsteps=1000,kT=1)
        works.append(work)
        avexpworks[j]+=np.exp(-work/kT)
    works=np.array(works)
    ## print(works)
    avworks[j]=np.mean(works)
    avexpworks[j]=avexpworks[j]/Ns[j]#np.mean(np.exp(-works/kT))

F=-kT*np.log(avexpworks)

plt.plot(Ns,F,'*r')
plt.title('av. free energy')#exp work')
plt.grid(axis='both')
plt.xlabel('N indep. trajectories')
plt.ylabel('free energy')
plt.xscale('log')
```

## Crook's theorem

```python
Nsteps=100
Ns=10000
delta=3
kT=1

avworksFW=np.zeros(1)
avF=np.zeros(1)

worksFW=[]
expworksFW=[]

for i in range(Ns):
    traj,work=MC3(delta=delta,nsteps=Nsteps,c0=0,bw=False)
    worksFW.append(work)
    expworksFW.append(np.exp(-work/kT))

worksFW=np.array(worksFW)

avF=-kT*np.log(np.mean(expworksFW))
print(avF)


plt.figure()
nbins=100
n, bins, patches = plt.hist(worksFW, nbins,alpha=0.5,density=True,label='forward')
print(patches)



worksBW=[]
expworksBW=[]

for i in range(Ns):
    traj,work=MC3(delta=delta,nsteps=Nsteps,c0=0,bw=True)
    worksBW.append(work)
    expworksBW.append(np.exp(-work/kT))

worksBW=np.array(worksBW)

avF=-kT*np.log(np.mean(expworksBW))
print(avF)

#plt.figure()
nbins=100
n, bins, patches = plt.hist(-worksBW, nbins,alpha=0.5,density=True,label='backward')
print(patches)
plt.xlabel('W')
plt.ylabel('probability density')
plt.grid(axis='both')
plt.show()
```

```python
plt.figure()
nbins=100
n1, bins, patches = plt.hist(worksFW, nbins,range=[0.325,0.350],alpha=0.5,density=True,label='forward')

n2, bins, patches = plt.hist(-worksBW, nbins,range=[0.325,0.350],alpha=0.5,density=True,label='backward')
plt.xlabel('W')
plt.ylabel('probability density')
plt.grid(axis='both')
plt.show()

print(n2-n1)
plt.plot(n2-n1,'*')
```
