We center and rescale the data to avoid problems when we perform dimensional reduction. To do so, we subtract from each feature its mean and we divide it for its standard deviation.

```python
@njit(fastmath=True)
def centering(x):
    """Function that centers and rescale the data"""
    y = np.zeros((x.shape))
    for i in range(x.shape[1]):
        y[:,i] = (x[:,i] - np.mean(x[:,i]))/(np.std(x[:,i]))
    
    return y
```

We import the function that we used in the previous assignment to perform PCA and we do it.

```python
def PCA(x):
    """ Function that evaluates the PCs of the given data"""
    # Evaluate covariance matrix
    Cov = np.cov(x, rowvar=False,bias=True)
    # Evaluate eigenvalues and eigenvectors
    w, A = np.linalg.eigh(Cov)
    # Flip the eigenvalues and eigenvectors, so that they are sorted from the biggest eigenvalue to the smallest
    w = np.flip(w)
    A = np.flip(A, axis=1)
    
    return w, A
```

In order to compare quantitavly the two distributions, we use the so called Mutual information.

```python
def mutual_information(Pxy):
    """ Function that evaluates the mutual information, given the joint probability"""
    I = 0.0
    Px = np.sum(Pxy, axis=1)
    Py = np.sum(Pxy, axis=0)
    for i in range(Pxy.shape[0]):
        for j in range(Pxy.shape[1]):
            if Pxy[i,j] != 0.0:
                I += Pxy[i,j]*np.log(Pxy[i,j]/(Px[i]*Py[j]))
                
    return I
```

# Kernel-PCA with gaussian Kernel

Now we try to do perform Kernel-PCA. First of all we evaluate the euclidean distance of all the pairs of points.

```python
D = euclidean_distances(x,x)
print(D.shape)
```

```python
print(D)
```

Now we sort all the columns of the euclidean matrix and we take the mean of the k-nearest neighbours.

```python
Dsort = np.copy(D)
for i in range(D.shape[0]):
    Dsort[:,i] = np.sort(Dsort[:,i])
    
width = []
for i in range(D.shape[0]):
    width.append(np.mean(Dsort[i,:]))

print(width[5])

del Dsort
```

We evaluate the matrix $K^{ij}$, using a Gaussian kernel. We use as parameter of the Gaussian kernel the fifth nearest-neighbour distance.

```python
K = np.exp(-0.5*(D/width[5])**2)
```

```python
print(K)
```

From $K^{ij}$ we do the double centering procedure:
\begin{equation}
    G{ij} = K^{ij} - \frac{1}{N} \sum_{l=1}^N \left(K^{il} + K^{jl} \right) + \frac{1}{N^2} \sum_{k,l=1}^N K^{kl} 
\end{equation}

```python
@njit(fastmath=True)
def double_centering(A):
    N = A.shape[0]
    G = np.zeros((N,N))
    Asum1 = np.zeros(N)
    for i in range(N):
        Asum1[i] = np.sum(A[:,i])
    Asum2 = np.sum(Asum1)
    for i in range(N):
        for j in range(i,N):
            G[i,j] = A[i,j] - 1.0/N*(Asum1[i]+Asum1[j]) + (1.0/N**2)*Asum2
            G[j,i] = G[i,j]
            
    return G
```
