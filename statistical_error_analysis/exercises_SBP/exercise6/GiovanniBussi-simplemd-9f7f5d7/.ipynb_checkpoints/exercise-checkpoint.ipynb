```python
%%bash

for dt in 0.001 0.002 0.005 0.01 0.02 0.05 0.1 0.2 0.5
do
echo $dt ## to print dt values
cat > in-$dt << EOF
inputfile xyz/crystal.xyz
outputfile xyz/output.xyz
temperature 0.722
tstep $dt
friction 0.0
forcecutoff 2.5
listcutoff 3.0
nstep $(awk -v dt=$dt 'BEGIN{print 20/dt}') ## to execute a command in bash
nconfig 10 xyz/trajectory-$dt.xyz
nstat 10 xyz/energies-$dt.dat
EOF
python3 python/simplemd.py < in-$dt & # & to execute more than one at the same time (CPU...)
done
```

# Find the max time step allowed such that the simulations do not explode for T=0.722, 256 atoms, up to max time = 20

```python
%%bash

for dt in 0.001 0.002 0.005 0.01 0.02 0.025 0.03 0.035 0.04 0.045 0.05 
do
echo $dt ## to print dt values
cat > in-$dt << EOF
inputfile xyz/crystal.xyz
outputfile xyz/output.xyz
temperature 0.722
tstep $dt
friction 0.0
forcecutoff 2.5
listcutoff 3.0
nstep $(awk -v dt=$dt 'BEGIN{print int(20/dt)}') ## to execute a command in bash
nconfig 10 xyz/trajectory-$dt.xyz
nstat 10 xyz/energies-$dt.dat
EOF
python3 python/simplemd.py < in-$dt & # & to execute more than one at the same time (CPU...)
done
```

```python
dts="0.001 0.002 0.005 0.01 0.02 0.025 0.03 0.035 0.04 0.045 0.05".split()
del dts[-1], dts[-1], dts[-1]
ene={} # a dictionary
for dt in dts:
    ene[dt]=np.loadtxt("xyz/energies-" + dt + ".dat").T
    

for dt in dts:
    plt.plot(ene[dt][1],ene[dt][4],label=dt)
plt.legend()
# plt.ylim((-1310,-1305))
```

# Linear fit and standard deviation: energy drift and fluctuations

```python
#for dt in dts:
dt="0.02"
np.polyfit(ene[dt][1],ene[dt][4],1)[0] # fit with a polynomial of order 1, the slope is the first (0) coefficient
```

```python
slopes=[]
flucts=[]
x=[]
for dt in dts[:-3]: # dts or dts[:-1] ## so excluding the last one
    x.append(float(dt))
    slopes.append(np.polyfit(ene[dt][1],ene[dt][4],1)[0])
    flucts.append(np.std(ene[dt][4]-slopes[-1]*ene[dt][1]))
print(slopes)
print(flucts)
plt.plot(x,slopes,'x') # since with plot(dts,slopes) the x variables are not ordered correctly 
plt.title('slopes')
plt.xscale('log')
#plt.yscale('log')
plt.show()

plt.title('fluctuations')
plt.plot(x,flucts,'o')
plt.xscale('log')
plt.yscale('log')
#legend('fluctuations')
```

## Now, let's try different values for the temperature

```python
%%bash
for T in 0.01 0.02 0.05 0.1 0.2 0.5 1.0 2.0 5.0
do

for dt in 0.001 0.002 0.005 0.01 0.02 0.05 0.1 0.2 0.5 1.0 2.0
do
echo $dt ## to print dt values
cat > in-$dt-$T << EOF
inputfile xyz/crystal.xyz
outputfile xyz/output.xyz
temperature $T
tstep $dt
friction 0.0
forcecutoff 2.5
listcutoff 3.0
nstep $(awk -v dt=$dt 'BEGIN{print 20/dt}') ## to execute a command in bash
nconfig 10 xyz/trajectory-$dt-$T.xyz
nstat 10 xyz/energies-$dt-$T.dat
EOF
python3 python/simplemd.py < in-$dt-$T & # & to execute more than one at the same time (CPU...)
done

done
```

```python
dts="0.001 0.002 0.005 0.01 0.02 0.05 0.1 0.2 0.5 1.0".split()
Ts="0.01 0.02 0.05 0.1 0.2 0.5 1.0 2.0 5.0".split()
# del dts[-1], dts[-1]#, dts[-1] ## but so doing, dts is modified; better for dt in dts[:-2]
ene={} # a dictionary
for dt in dts:
    ene[dt]=np.loadtxt("xyz/energies-" + dt + "-0.01.dat").T
    ## then change the value for the temperature in the output plots
    

for dt in dts:
    plt.plot(ene[dt][1],ene[dt][4],label=dt)
plt.legend()
plt.yscale('log')
# plt.ylim((-1310,-1305))
```

```python
slopes=[]
flucts=[]
x=[]
for dt in dts: # dts or dts[:-1] ## so excluding the last one
    x.append(float(dt))
    slopes.append(np.polyfit(ene[dt][1],ene[dt][4],1)[0])
    flucts.append(np.std(ene[dt][4]-slopes[-1]*ene[dt][1]))
print(slopes)
print(flucts)
plt.plot(x,slopes,'o') # since with plot(dts,slopes) the x variables are not ordered correctly 
plt.title('slopes')
plt.xscale('log')
#plt.yscale('log')
plt.show()

plt.title('fluctuations')
plt.plot(x,flucts,'o')
plt.xscale('log')
plt.yscale('log')
#legend('fluctuations')
## WATCH OUT: in the estimate of fluctuations there is an error due to very big fluctuations at first few times (overestimate)
## (see the plots for the N of particles below); do ene[dt][4][20:], i.e. consider ene from (NOTED) iteration 20
```

```python
print(x)
```

```python
## however, it's better to have all the computed values for different temperatures in a unique "array"
ene2={}
for T in Ts:
    for dt in dts:
        ene2[T,dt]=np.loadtxt("xyz/energies-" + dt + "-" + T + ".dat").T        
```

```python
slopes=[]
flucts=[]
x=[]
for k in ene2.keys(): # dts or dts[:-1] ## so excluding the last one
    x.append(k)
    slopes.append(np.polyfit(ene2[k][1],ene2[k][4],1)[0])
    flucts.append(np.std(ene2[k][4]-slopes[-1]*ene[k][1]))
```

```python
for i in range(len(x)):
    print(x[i],slopes[i])
```

increasing temperature, the maximum time step is lower

# Next, let's try for different number of particles
To see how the properties depend on the system size (constant density)

```python
%%bash
for N in 1 2 3 4 5 6 7 8
do
python3 xyz/lattice.py $N > xyz/crystal-$N.xyz
done
```

```python
%%bash
ls xyz -ltr
```

```python
%%bash
for N in 1 2 3 4 5 6 7 8
do

cat > in-$N << EOF

inputfile xyz/crystal-$N.xyz
outputfile xyz/output.xyz
temperature 0.722
tstep 0.005
friction 0.0
forcecutoff 2.5
listcutoff 3.0
nstep 4000
nconfig 10 xyz/trajectory-$N.xyz
nstat 10 xyz/energies-$N.dat
EOF

python3 python/simplemd.py < in-$N &

done
```

```python
Ns="1 2 3 4 5 6 7 8".split()
```

```python
ene3={}
for N in Ns:
    ene3[N]=np.loadtxt("xyz/energies-" + N + ".dat").T
    
```

```python
ene3["1"]
```

```python
slopesN=[]
fluctsN=[]
xN=[]
##for k in ene3.keys(): # dts or dts[:-1] ## so excluding the last one
for N in Ns:
    xN.append(4*float(N)**3)
    slopesN.append(np.polyfit(ene3[N][1],ene3[N][4],1)[0])
    fluctsN.append(np.std(ene3[N][4]-slopesN[-1]*ene3[N][1]))
```

```python
for i in range(len(xN)):
    print(xN[i],slopesN[i],fluctsN[i])
```

```python
plt.plot(xN,slopesN,'o')
plt.title('slope')
plt.show()
plt.plot(xN,fluctsN,'x-')
plt.title('fluctuations')
```

except for very few atoms, the slope is small and the fluctuations (i.e. the standard deviation) are increasing,
however I expected the std to increase as the square root of the system size (variances are additive on the different
parts of the system)

```python
plt.plot(xN,np.array(fluctsN)/xN,'x-') # array(fluctsN)/np.sqrt(xN) is not constant
plt.ylim((-0.001,0.005))
```

Why it does not seem to happen? Let's see ener3: there are very big fluctuations at the first few times, giving a wrong estimate for fluctuations

```python
plt.plot(ene3["5"][4])
```

```python
plt.plot(ene3["5"][4][20:])
```

```python
slopesN=[]
fluctsN=[]
xN=[]
##for k in ene3.keys(): # dts or dts[:-1] ## so excluding the last one
for N in Ns:
    xN.append(4*float(N)**3)
    slopesN.append(np.polyfit(ene3[N][1][20:],ene3[N][4][20:],1)[0])
    fluctsN.append(np.std(ene3[N][4][20:]-slopesN[-1]*ene3[N][1][20:]))
```

```python
plt.plot(xN,slopesN,'o')
plt.title('slope')
plt.show()
plt.plot(xN,fluctsN,'x-')
plt.title('fluctuations')
```

```python
plt.plot(xN,np.array(fluctsN)/np.sqrt(xN),'x-')
#plt.ylim((-0.001,0.001))
plt.yscale('log')
## so, for smaller systems a small time step (big fluctuations), then constant value for the max time step
## since fluctuations are almost constant (with the system size) -- the energy non-conservation is to be valued per particle
```
