## coretools

```python
import coretools
```

```python
class MytoolResult(coretools.Result):
    """Result of a mytool calculation."""
    pass

def mytool():
    a = 3
    b = "ciao"
    return MytoolResult(a=a, b=b)

m = mytool()
print(m)
```

## Metropolis

```python
from Functions_mine import run_Metropolis
```

```python
x0 = 1

def my_energy_fun(x0):
    return x0**2

my_energy = {'fun': my_energy_fun, 'args': ()}

def my_proposal_fun(x0):
    
    r = np.random.randint(0, 2, 1)
    if r == 0: x_new = x0 - 0.01
    else: x_new = x0 + 0.01
    
    return x_new

my_proposal = {'fun': my_proposal_fun, 'args': ()}


out = run_Metropolis(x0, proposal=my_proposal, energy_function=my_energy)
```

```python
plt.plot(out[0])
```
