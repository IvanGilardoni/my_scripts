# Stochastic differential equations
1. Write a program to integrate the following stochastic differential equation:
\begin{equation}
dx = −xdt + \sqrt 2 dW
\end{equation}

Use Euler integrator (i.e. just replace dt with ∆t and dW with sqrt{∆t}R where R is a Gaussian number with zero average and unitary variance; choose a short timestep ∆t=0.001).

```python
import numpy as np
import matplotlib.pyplot as plt
from numba import njit
from scipy.optimize import curve_fit
```

```python
njit(fastmath=True) ## why it gives me error?
def run(dt,N,q=0):#,seed=0):
    np.random.seed(seed)
    traj=[]
    traj.append(q)
    
    for i in range(N):
        q=q*(1-dt)+np.sqrt(2*dt)*np.random.normal()
        traj.append(q)
        # traj.append((i*dt),q)
    return np.array(traj)
    
```

```python
dt=0.001
Tmax=10
q0=0
N=round(Tmax/dt)

time=np.linspace(0,Tmax,N+1)
traj=run(dt,N,q0)
```

```python
plt.plot(time,traj)
plt.xlabel('time')
plt.ylabel('trajectory')
plt.grid(axis='both')
```

2. Using this program, compute ten trajectories with same initial condition
(x = 10) and different seeds for the random number generator. At fixed
value of time t, compute the average over the trajectories of the value of
x and its standard deviation. How do these two quantities depend on t?

```python
M=100 # n. of trajectories
Tmax=20
dt=10**(-3)
N=round(Tmax/dt)
traj=np.zeros((N+1,M)) # N+1 due to initial position
time=np.linspace(0,Tmax,N+1)
q0=10

for j in range(M):
    traj[:,j]=run(dt,N,q0) # seed=itraj

avtraj=np.mean(traj,axis=1)
stdtraj=np.std(traj,axis=1)
```

```python
#for j in range(M):
    #plt.plot(traj[:,j],'-',label=j)

f=plt.figure(1)    
plt.plot(time,avtraj,'-',label='average')
#plt.plot(time,stdtraj,'-',label='std')
plt.plot(time,avtraj+stdtraj,'-',label='av+std')
plt.plot(time,avtraj-stdtraj,'-',label='av-std')

plt.legend()
plt.xlabel('time')
plt.ylabel('trajectories')
plt.grid(axis='both')

f2=plt.figure(2)
plt.plot(time,stdtraj**2,'-',label='std')
plt.xlabel('time')
plt.ylabel('variance')
plt.grid(axis='both')
```

```python
# <x>=e^{-t} (Langevin) expected, so fit average with this function
def funcexp(x,a,b,c):
    return a*np.exp(-b*x)+c

popt,pcov=curve_fit(funcexp,time,avtraj,[q0,1,0]) # popt optimal parameters, pcov the covariance matrix

plt.figure(figsize=(10,6))

plt.plot(time, funcexp(time, popt[0], popt[1],popt[2]),'-g',linewidth=3,label='fit')
plt.plot(time,avtraj,'-b',label='average')
plt.plot(time,avtraj+stdtraj,'-r',label='av+/-std')
plt.plot(time,avtraj-stdtraj,'-r')#,label='av-std')

plt.grid(axis='both')
plt.xlabel('time')
plt.ylabel('trajectory')
plt.legend()

print(popt)
print(pcov)
```

```python
# sigma^2(t)=1-e^(-2t)
def funcexp2(x,a,b):
    return a*(np.exp(-b*x)-1)

popt,pcov=curve_fit(funcexp2,time,stdtraj**2,[-1,2]) # popt optimal parameters, pcov the covariance matrix

plt.figure(figsize=(10,6))

plt.plot(time, funcexp2(time, popt[0], popt[1]),'-b',linewidth=3,label='fit')
plt.plot(time,stdtraj**2,'-r',label='variance')

plt.grid(axis='both')
plt.xlabel('time')
plt.ylabel('variance')
plt.legend()

print(popt)
print(pcov)
```

3. Analyze the trajectories and compute their time increment, defined as g(t) = x(t + τ ) − x(t).

How does the increment depend on τ ? How does it depend on x?

```python
n=[1,2,5,10,20]#,50,100,200,500,1000,2000,5000]

plt.figure(figsize=(10,6))

for i in range(np.size(n)):
    tau=n[i]*dt
    deltas=np.zeros(N-n[i])
    for j in range(N-n[i]):
        deltas[j]=avtraj[j+n[i]]-avtraj[j]
    plt.plot(deltas,'o',label='tau=%.3f' %(n[i]*dt),markersize=4)

plt.legend()
plt.xlim([0,1000])
plt.grid(axis='both')
plt.xlabel('t')
plt.ylabel('g(t)')
#newtraj=np.delete(traj,index,axis=0)

#np.difference(newtraj-traj,axis=0)
```
