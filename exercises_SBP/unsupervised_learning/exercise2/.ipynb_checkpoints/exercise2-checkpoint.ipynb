### 1. Preprocess the data in the first file by centering the variables and divide them by their standard deviation

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from numba import njit
import os.path
import time
from sklearn.linear_model import LogisticRegression
from sklearn.metrics.pairwise import euclidean_distances
# from scipy.linalg import eigh
from scipy.sparse.linalg import eigsh
```

```python
x = np.loadtxt('data/data_kPCA.txt', skiprows=0, delimiter=' ', usecols=tuple(np.arange(0, 10, 1)), unpack=True).T
label = np.loadtxt('data/labels_kPCA.txt', skiprows=0)

size1,size2=x.shape
print(size1)
print(size2)
print(np.min(label))
print(np.max(label))
```

```python
xmean=np.zeros(size2)
xstd=np.zeros(size2)

for i in range(size2):
    xmean[i]=np.mean(x[:,i])
    xstd[i]=np.std(x[:,i])

print('means: ',xmean)
print('stds: ',xstd)

x=(x-xmean)/xstd
```

### 2. Divide your data set into two. One (with the first 1000 data points and labels) would be employed as test set, while the other (the last 9000 data points) would be employed as learning set.

```python
size3=1000

# shuffle:
randomize=np.arange(size1)
np.random.shuffle(randomize)
x=x[np.argsort(randomize),:]
label=label[np.argsort(randomize)]
del randomize

testdata=x[0:size3,0:10]
testlabel=label[0:size3]
learndata=x[size3:10000,0:10]
learnlabel=label[size3:10000]
```

### 4. For an increasing number of principal components (from 1 to 10):
(a) Apply a multinomial logistic regression (you can program it yourself

(b) Transform the coordinates of the test data set with the matrix learned

```python
y=np.zeros((size1-size3,size2))
xPCs=np.zeros((size3,size2))
mutinf=np.zeros(size2)
acc=np.zeros(size2)

for i in range(size1-size3):
    for j in range(size2):
        y[i,j]=np.sum(learndata[i,:]*eigvecs[:,j])
        
for i in range(size3):
    for j in range(size2):
        xPCs[i,j]=np.sum(testdata[i,:]*eigvecs[:,j])
```

```python
for j in range(size2):
    clf = LogisticRegression(random_state=0,max_iter=1000).fit(y[:,0:j+1], learnlabel)
    #coefs=clf.coef_
    #print(coefs)
    
    pred=clf.predict(xPCs[:,0:j+1]) # prediction for the test set
    acc[j]=clf.score(xPCs[:,0:j+1], testlabel) # accuracy = correct predictions / total n.
    mutinf[j]=sklearn.metrics.mutual_info_score(testlabel,pred)
        
print(mutinf)
print(acc)
```

### 5. Repeat points 3. and 4. but using the kernel-PCA with a Gaussian kernel. Use as width, for instance, the average distance from the fifth nearest neighbor for each data point. Consider the entire data set for kernel PCA

```python
@njit(fastmath=True)
### Gram=kernel-1.0/size1*(np.sum(kernel,axis=0)+np.sum(kernel,axis=1))+1/(2*size1**2)*np.sum(kernel)
def double_centering(kernel):
    sum1=np.zeros(size1)
    Gram=np.zeros((size1,size1))
    for i in range(size1):
        sum1[i]=np.sum(kernel[:,i])
    sum2=np.sum(sum1)
    for i in range(size1):
        for j in range(i,size1):
            Gram[i,j]=kernel[i,j]-1.0/size1*(sum1[i]+sum1[j])+(1.0/size1**2)*sum2
            Gram[j,i]=Gram[i,j]      

    return(Gram)
```

in the following, you are doing this step by step for a chosen value of NN;

otherwise, you can skip to the implemented function to study (one or) different values for the NN

```python
distances=euclidean_distances(x,x)
print(distances)
distances2=distances**2
distances.sort(axis=1)

NN=10 # nearest neighbor for the width of the gaussian
width=np.mean(distances[:,NN+1]) ## +1 since it starts from zero

del distances # delete distances (since now it is sorted)

kernel=np.exp(-distances2/(2*width**2))
```

### 6) Implement ISOMAP, project the data set into 2D and color by label.
the idea is: performing MDS using as a distance an estimate of the geodetic distance in the embedded manifold

```python
@njit(fastmath=True)
def geodetic(size1,distances):
    for k in range(size1):
        if (k%500==0):
            print(k)
        for i in range(size1):
            for j in range(size1):
                if distances[i,j]>distances[i,k]+distances[k,j]:
                    distances[i,j]=distances[i,k]+distances[k,j]
    return distances
```
