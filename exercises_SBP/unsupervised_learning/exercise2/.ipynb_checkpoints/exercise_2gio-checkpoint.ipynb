```python
# Importing dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
# Importing the dataset
df_X = pd.read_csv("Unsupervised-Learning-Datasets-main/data_kPCA.txt", sep=" ", header=None).drop(10,axis=1)
X = df_X.values
df_X.head()
```

```python
df_y = pd.read_csv("Unsupervised-Learning-Datasets-main/labels_kPCA.txt", sep=" ", header=None)
df_y.columns = ["label"]
y = df_y.values
df_y.head()
```

```python
# Concatenating the two dataframes
df = pd.concat([df_X, df_y], axis=1)
df.head()
```

## 1. Data Preprocessing
Preprocess the data in the first file by centering the variables and divide them by their standard deviation.

```python
# Standardizing dataset
std = X.std(axis=0)
mean = X.mean(axis=0)
X_std = (X-mean)/std
```

## 2. Train-test split
Divide your data set into two. One (with the first 1000 data points and labels) would be employed as test set, while the other (the last 9000 data points) would be employed as learning set.

SInce it's not something special, I do it using scikit-learn implementation.

```python
from sklearn.model_selection import train_test_split
```

```python
test_size = 0.1  # fraction of elements in test set
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size=test_size, random_state=42)
y_train = y_train.reshape(X_train.shape[0],)
y_test = y_test.reshape(X_test.shape[0],)
```

## 3. PCA
Use the program from the previous exercise for computing the Principal Components Analysis on the learning data set. Obtain and plot the eigenvalue spectrum using logscale for the y-axis. Project the data in the two first PCs and color it by label.

```python
# Coviariance matrix applying definition (using matrix multiplication)
N_train = X_train.shape[0] # number of points
n_features = X_train.shape[1]  # number of features

# C = X_train.transpose() @ X_train /N_train
C = np.cov(X_train.transpose())
```

```python
# Compute eigenvalues
eig_values, eig_vectors = np.linalg.eig(C)
# Sorting eigenvalues and eigenvectors
index_sort = np.argsort(eig_values)[::-1]  # getting indexes for sorting (eigenvalues in descending order)
eig_values = eig_values[index_sort]
eig_vectors = eig_vectors[index_sort]
```

```python
eig_vectors.shape
```

```python
# Plotting eigenvalues
plt.subplots(figsize=(9,6))
plt.plot(eig_values,'.',markersize=15)
plt.title('Eigenvalues spectrum', fontsize=15)
plt.ylabel(r'$\lambda$', fontsize=18)
plt.yscale('log')
```

### Plot of the first two components

```python
# Projecting all the dataset onto the PCs space
X_train_pca =  X_train @ eig_vectors
X_test_pca =   X_test @ eig_vectors
```

```python
# creating dataframe with projected features
df_train_pca = pd.DataFrame(X_train_pca, columns=["PC"+str(i+1) for i in range(n_features)])
df_train_pca[["label"]] = y_train.reshape((N_train,1))
df_train_pca.head()
```

```python
# Plot first two PCs, coloring data points by family
plt.subplots(figsize=(10,7))
sns.scatterplot(x="PC1", y="PC2", data=df_train_pca, hue="label")
plt.ylim(-7,8)
plt.show()
```

## 4. Multinomial Logistic Regression
For an increasing number of principal components (from 1 to 10):
1. Apply a multinomial logistic regression (you can program it yourself or use an external library) for learning a model in the learning dataset.
2. Transform the coordinates of the test data set with the matrix learned from the learning data set and make a prediction based on the logistic model. Quantify the quality of the prediction by computing the Mutual Information between the ground truth classification and the predicted labels.

I use the scikit-learn implementation for multinomial logistic regression.

```python
from sklearn.linear_model import LogisticRegression
```

```python
def mutual_information(y_test,y_pred):
    """
    Function to compute mutual information between two discrete arrays. In our case the are the predicted labels and the correct ones.
    
    Input:
    - y_test: array with ground-truth labels
    - y_pred: array with predicted labels
    
    Output:
    - mi: discrete mutual information between the two arrays
    
    Another possible way to compute mutual information is to use scikit-learn implementation "mutual_info_classif":
    from sklearn.feature_selection import mutual_info_classif
    mic = mutual_info_classif(y_test.reshape(-1,1),y_pred.reshape(-1,), discrete_features=True)
    
    """
    labels = np.unique(y_train)
    N_test = len(y_test)
    mi = 0  # initializing mutual information
    for y in labels:  # loop over possible labels
        for x in labels:
            p_xy = sum((y_test==x) & (y_pred==y))/N_test  # p(x,y)
            p_x = sum((y_test==x))/N_test  # p(x)
            p_y = sum((y_pred==y))/N_test  # p(y)
            if p_xy!=0:  # if p(x,y)==0 then p(x,y)*log(p(x,y))=0 and mi remains the same
                mi = mi + p_xy * np.log(p_xy/(p_x*p_y))
                
    return mi
```

```python
n_pc_vec = np.linspace(1,10,10, dtype=int)  # number of principal components used for training
mi_vec = np.zeros(len(n_pc_vec))
for i,n_pc in enumerate(n_pc_vec):
    mlr = LogisticRegression(random_state=0, max_iter=1000)
    mlr.fit(X_train_pca[:,0:n_pc], y_train)  # fitting the model with the first "n_pc" principal components
    y_pred = mlr.predict(X_test_pca[:,0:n_pc])
    mi_vec[i] = mutual_information(y_test, y_pred)
```

```python
# Plot mutual information as a function of the number of components
plt.subplots(figsize=(8,5))
plt.plot(n_pc_vec, mi_vec, '.-')
plt.xlabel("# of PCs for training")
plt.ylabel("Mutual information")
```

## 5. Kernel-PCA
Repeat points 3. and 4. but using the kernel-PCA with a Gaussian kernel. Use as width, for instance, the average distance from the fifth nearest neighbor for each data point.

```python
# I compute the average distance from the 5-th nearest neighbor (in the training set). As distance I use the classical euclidean distance

# # Trivial implementation (slow)
# N_train = X_train.shape[0]
# dist = np.zeros((N_train,N_train))
# for i in range(N_train):
#     # computing the distance betwee i and all the other datapoints
#     for j in range(i,N_train):
#         dist[i,j] = sum((X_train[i,:] - X_train[j,:])*(X_train[i,:] - X_train[j,:]))
#         dist[j,i] = dist[i,j]

# dist = np.sqrt(dist)  # taking the square root

# I do it using scipy implemented function since it is way faster than my implementation
from scipy.spatial import distance_matrix
dist = distance_matrix(X_std, X_std, p=2)
```

```python
# now I compute the average distance from the 5-th nearest neighbor of each datapoint
dist5_mean = 0
N = X_std.shape[0]
for i in range(N):
    dist_ordered = np.sort(dist[i,:])
    # dist_ordered[5] is the distance to the fifth nearest-neighbor since the array includes the distance with itslef (which is zero)
    dist5_mean = dist5_mean + dist_ordered[5]

dist5_mean = dist5_mean/N
```

```python
# Defining the kernel
sigma = 5*dist5_mean  # sigma equal to dist5_mean is too small
K = np.exp(-(dist**2)/(2*(sigma**2)))  # kernel matrix
```

```python
# I compute GRAM matrix with DOUBLE CENTERING formula
# sum_K = np.sum(K)/(N_train**2)
# sum_row = np.sum(K, axis=0)/N_train
# G = np.zeros((N_train,N_train))
# for i in range(N_train):
#     # computing the distance betwee i and all the other datapoints
#     for j in range(i,N_train):
#         G[i,j] = K[i,j] + sum_K - sum_row[i] - sum_row[j]
#         G[j,i] = G[i,j]
## or equivalently
I = np.identity(N)
J_N = np.ones(N)/N
G = (I-J_N) @ K @ (I-J_N)  

# Equivalently one can compute the GRAM matrix using scikit-learn implementation --> exact same result but way faster
from sklearn.preprocessing import KernelCenterer
kern_cent = KernelCenterer()
G_skl = kern_cent.fit_transform(K)
```

```python
# Computing the first 10 eigenvalues and eigenvectors of G
from scipy.sparse.linalg import eigsh
eig_values, eig_vectors = eigsh(G_skl,10)
eig_values = np.real(eig_values)
eig_vectors = np.real(eig_vectors)
eig_values
```

```python
# Sorting eigenvalues and eigenvectors
index_sort = np.argsort(eig_values)[::-1]  # getting indexes for sorting (eigenvalues in descending order)
eig_values = eig_values[index_sort]
eig_vectors = eig_vectors[:,index_sort]
eig_vectors = eig_vectors*np.sqrt(eig_values)
eig_vectors.shape
```

```python
X_kpca = eig_vectors.copy()
# creating dataframe with projected features
df_kpca = pd.DataFrame(X_kpca, columns=["kPC"+str(i+1) for i in range(n_features)])
df_kpca[["label"]] = y.reshape((N,1))
df_kpca.head()
```

```python
# Plotting eigenvalues
plt.subplots(figsize=(9,6))
plt.plot(eig_values,'.',markersize=15)
plt.title('Eigenvalues spectrum', fontsize=15)
plt.ylabel(r'$\lambda$', fontsize=18)
```

```python
# Plot first two PCs, coloring data points by family
plt.subplots(figsize=(10,7))
sns.scatterplot(x="kPC1", y="kPC2", data=df_kpca, hue="label")
```

### Multinomial Logistic Regression

```python
X_train_kpca, X_test_kpca, y_train_kpca, y_test_kpca = train_test_split(X_kpca, y, test_size=test_size, random_state=42)
y_train_kpca = y_train_kpca.reshape(X_train_kpca.shape[0],)
y_test_kpca = y_test_kpca.reshape(X_test_kpca.shape[0],)
```

```python
n_pc_vec = np.linspace(1,10,10, dtype=int)  # number of principal components used for training
mi_vec = np.zeros(len(n_pc_vec))
for i,n_pc in enumerate(n_pc_vec):
    mlr = LogisticRegression(random_state=0, max_iter=1000)
    mlr.fit(X_train_kpca[:,0:n_pc], y_train_kpca)  # fitting the model with the first "n_pc" components
    y_pred = mlr.predict(X_test_kpca[:,0:n_pc])
    mi_vec[i] = mutual_information(y_test_kpca, y_pred)
```

```python
# Plot mutual information as a function of the number of components
plt.subplots(figsize=(8,5))
plt.plot(n_pc_vec, mi_vec, '.-')
plt.xlabel("# of PCs for training")
plt.ylabel("Mutual information")
```
