### 1. Preprocess the data in the first file by centering the variables and divide them by their standard deviation

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from numba import njit
import os.path
import time
from sklearn.linear_model import LogisticRegression
from sklearn.metrics.pairwise import euclidean_distances
# from scipy.linalg import eigh
from scipy.sparse.linalg import eigsh
```

```python
x = np.loadtxt('data/data_kPCA.txt', skiprows=0, delimiter=' ', usecols=tuple(np.arange(0, 10, 1)), unpack=True).T
label = np.loadtxt('data/labels_kPCA.txt', skiprows=0)

size1,size2=x.shape
print(size1)
print(size2)
print(np.unique(label))
```

```python
xmean=np.zeros(size2)
xstd=np.zeros(size2)

for i in range(size2):
    xmean[i]=np.mean(x[:,i])
    xstd[i]=np.std(x[:,i])

print('means: ',xmean)
print('stds: ',xstd)

x=(x-xmean)/xstd
```

### 2. Divide your data set into two. One (with the first 1000 data points and labels) would be employed as test set, while the other (the last 9000 data points) would be employed as learning set.

```python
size3=1000

# shuffle:
randomize=np.arange(size1)
np.random.shuffle(randomize)
x=x[np.argsort(randomize),:]
label=label[np.argsort(randomize)]
del randomize

testdata=x[0:size3,0:10]
testlabel=label[0:size3]
learndata=x[size3:10000,0:10]
learnlabel=label[size3:10000]
```

### 3.Use the program from the previous exercise for computing the Principal Components Analysis on the learning data set. Obtain and plot the eigenvalue spectrum using logscale for the y-axis. Project the data in the two first PCs and color it by label.

```python
cov=np.cov(learndata.T)
eigvals,eigvecs=np.linalg.eig(cov) # or eigh

idx=eigvals.argsort()[::-1]   
eigvals=eigvals[idx]
eigvecs=eigvecs[:,idx]

neigvals=eigvals/sum(eigvals) # "normalized" eigenvalues
print('"normalized" eigenvalues: ',neigvals)
print('eigvenvalues: ',eigvals)

plt.plot(neigvals,'*')
plt.yscale('log')
plt.xlabel('component n.')
plt.ylabel('eigenvalues')
plt.grid(axis='both')
```

```python
y1=np.zeros(size1-size3)
y2=np.zeros(size1-size3)

for i in range(size1-size3):
    y1[i]=np.sum(learndata[i,:]*eigvecs[:,0])
    y2[i]=np.sum(learndata[i,:]*eigvecs[:,1])
```

```python
f=plt.figure(figsize=(10,7))

plt.title("PCs projection (training set)", fontsize=20)

plt.xlabel("PC 1", fontsize=15)
plt.ylabel("PC 2", fontsize=15)

plt.plot(y1[np.where(learnlabel==0)],y2[np.where(learnlabel==0)],'o',label='0',markersize='2')
plt.plot(y1[np.where(learnlabel==1)],y2[np.where(learnlabel==1)],'o',label='1',markersize='2')
plt.plot(y1[np.where(learnlabel==2)],y2[np.where(learnlabel==2)],'o',label='2',markersize='2')
plt.plot(y1[np.where(learnlabel==3)],y2[np.where(learnlabel==3)],'o',label='3',markersize='2')

plt.grid(axis='both')
plt.legend()

plt.show()

```

```python
x1=np.zeros(size3)
x2=np.zeros(size3)

for i in range(size3):
    x1[i]=np.sum(testdata[i,:]*eigvecs[:,0])
    x2[i]=np.sum(testdata[i,:]*eigvecs[:,1])

f=plt.figure(figsize=(10,7))

plt.title("PCs projection (test set)", fontsize=20)

plt.xlabel("PC 1", fontsize=15)
plt.ylabel("PC 2", fontsize=15)

plt.plot(x1[np.where(testlabel==0)],x2[np.where(testlabel==0)],'o',label='0')
plt.plot(x1[np.where(testlabel==1)],x2[np.where(testlabel==1)],'o',label='1')
plt.plot(x1[np.where(testlabel==2)],x2[np.where(testlabel==2)],'o',label='2')
plt.plot(x1[np.where(testlabel==3)],x2[np.where(testlabel==3)],'o',label='3')

plt.grid(axis='both')
plt.legend()

plt.show()
```

### 4. For an increasing number of principal components (from 1 to 10):
(a) Apply a multinomial logistic regression (you can program it yourself
or use an external library) for learning a model in the learning data
set.


(b) Transform the coordinates of the test data set with the matrix learned
from the learning data set and make a prediction based on the logistic model. Quantify the quality of the prediction by computing the
Mutual Information between the ground truth classication and the
predicted labels.

```python
y=np.zeros((size1-size3,size2))
xPCs=np.zeros((size3,size2))
mutinf=np.zeros(size2)
acc=np.zeros(size2)

for i in range(size1-size3):
    for j in range(size2):
        y[i,j]=np.sum(learndata[i,:]*eigvecs[:,j])
        
for i in range(size3):
    for j in range(size2):
        xPCs[i,j]=np.sum(testdata[i,:]*eigvecs[:,j])
```

```python
for j in range(size2):
    clf = LogisticRegression(random_state=0,max_iter=1000).fit(y[:,0:j+1], learnlabel)
    
    pred=clf.predict(xPCs[:,0:j+1]) # prediction for the test set
    acc[j]=clf.score(xPCs[:,0:j+1], testlabel) # accuracy = correct predictions / total n.
    mutinf[j]=sklearn.metrics.mutual_info_score(testlabel,pred)
        
print(mutinf)
print(acc)
```

```python
[hist,hist2]=np.histogram(testlabel,bins=[0,1,2,3,4]) # bins: [0,1) and so on
hist=hist/np.sum(hist)
entropy=-np.sum(hist*np.log(hist))
del hist, hist2
print('entropy =',entropy)

plt.plot(mutinf,'bo',label='mutual information')
plt.hlines(entropy,0,10,'r','dashed',label='entropy')
plt.xlabel("n PCs", fontsize=15)
plt.ylabel("mutual information", fontsize=15)
plt.legend(loc='center right')
plt.grid(axis='both')

plt.figure()
plt.plot(acc,'ro')
plt.xlabel("n PCs", fontsize=15)
plt.ylabel("accuracy", fontsize=15)
plt.ylim([0.6,1])
plt.title('accuracy')
plt.grid(axis='both')
```

### 5. Repeat points 3. and 4. but using the kernel-PCA with a Gaussian kernel. Use as width, for instance, the average distance from the fifth nearest neighbor for each data point. Consider the entire data set for kernel PCA

```python
@njit(fastmath=True)
### Gram=kernel-1.0/size1*(np.sum(kernel,axis=0)+np.sum(kernel,axis=1))+1/(2*size1**2)*np.sum(kernel)
def double_centering(kernel):
    sum1=np.zeros(size1)
    Gram=np.zeros((size1,size1))
    for i in range(size1):
        sum1[i]=np.sum(kernel[:,i])
    sum2=np.sum(sum1)
    for i in range(size1):
        for j in range(i,size1):
            Gram[i,j]=kernel[i,j]-1.0/size1*(sum1[i]+sum1[j])+(1.0/size1**2)*sum2
            Gram[j,i]=Gram[i,j]      

    return(Gram)
```

```python
nNNs=[5,10,50,100,500,1000]
k=10 # nPCs=size2=10
bplot=1

Keigvalss=np.zeros((np.size(nNNs),k))
Kmutinfs=np.zeros((np.size(nNNs),k))
Kaccs=np.zeros((np.size(nNNs),k))
    
# 1. compute the average distance to the nNN nearest neighbour
distances=euclidean_distances(x,x)
distances2=distances**2
distances.sort(axis=1) # sort distances (only to compute the n-th NN)


i=0
for nNN in nNNs:
    width=np.mean(distances[:,nNN])

    # 2. compute the kernel and the Gram matrix, diagonalize it
    kernel=np.exp(-distances2/(2*width**2))
    Gram=double_centering(kernel)
    Keigvals, Keigvecs = eigsh(Gram, k, which='LA')

    idx=Keigvals.argsort()[::-1]   
    Keigvals=Keigvals[idx]
    Keigvecs=Keigvecs[:,idx]

    # 3. optional plot of the highest k eigenvalues
    if (bplot==1):
        print('\n\n N OF NEIGHBOURS: %i'%nNN)
        print('width= ',width)
        plt.title('eigenvalues')
        plt.plot(Keigvals,'*')
        plt.xlabel('component n.')
        plt.ylabel('eigenvalues')
        plt.grid(axis='both')

    # 4. optional plot of the projection into the highest 2 eigenvectors
    Keig0=Keigvecs[:,0]
    Keig1=Keigvecs[:,1]

    if (bplot==1):
        f=plt.figure(figsize=(10,7))
        plt.title("Kernel PCs projection", fontsize=20)

        plt.xlabel("K PC 1", fontsize=15)
        plt.ylabel("K PC 2", fontsize=15)

        plt.plot(Keig0[np.where(label==0)],Keig1[np.where(label==0)],'o',label='0',markersize='2')
        plt.plot(Keig0[np.where(label==1)],Keig1[np.where(label==1)],'o',label='1',markersize='2')
        plt.plot(Keig0[np.where(label==2)],Keig1[np.where(label==2)],'o',label='2',markersize='2')
        plt.plot(Keig0[np.where(label==3)],Keig1[np.where(label==3)],'o',label='3',markersize='2')

        plt.grid(axis='both')
        plt.legend()
        plt.show()

    # 5. logistic regression: mutual information and accuracy
    Kmutinf=np.zeros(k)
    Kacc=np.zeros(k)

    for j in range(k):
    
        Kclf = LogisticRegression(random_state=0,max_iter=1000).fit(Keigvecs[size3:10000,0:j+1], learnlabel)
    
        Kpred=Kclf.predict(Keigvecs[0:size3,0:j+1])
        Kmutinf[j]=sklearn.metrics.mutual_info_score(testlabel,Kpred)
        Kacc[j]=Kclf.score(Keigvecs[0:size3,0:j+1], testlabel)

    if (bplot==1):
        print('eigenvalues: ',Keigvals)
        print('mutual information: ',Kmutinf)
        print('accuracy: ',Kacc)

    [Keigvalss[i,:],Kmutinfs[i,:],Kaccs[i,:]]=[Keigvals,Kmutinf,Kacc]
    i+=1

del distances # because now it is sorted
```

```python
# plots of: eigenvalues, mutual information, accuracy
nkPCs=np.arange(1,k+1)
nPCs=np.arange(1,11)

plt.figure()
plt.title('Eigenvalues vs. n PCs', fontsize=15)
plt.xlabel("n PCs", fontsize=15)
plt.ylabel("eigenvalues", fontsize=15)

for i in range(np.size(nNNs)):
    plt.plot(nkPCs,Keigvalss[i,:],'--o',label='gkPCA, NN=%i'%nNNs[i])

plt.grid(axis='both')
plt.legend()
plt.show()



plt.figure()
plt.title('Mutual information vs. n PCs', fontsize=15)
plt.xlabel("n PCs", fontsize=15)
plt.ylabel("mutual information", fontsize=15)


for i in range(np.size(nNNs)):
    plt.plot(nkPCs,Kmutinfs[i,:],'--o',label='gkPCA, NN=%i'%nNNs[i])
plt.plot(nPCs,mutinf,'--o',label='PCA')
plt.hlines(entropy,0,10,'r','dashed',label='entropy')

plt.grid(axis='both')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()


plt.figure()
plt.title('Accuracy vs. n PCs', fontsize=15)
plt.xlabel("n PCs", fontsize=15)
plt.ylabel("accuracy", fontsize=15)

for i in range(np.size(nNNs)):
    plt.plot(nkPCs,Kaccs[i,:],'--o',label='gkPCA, NN%i'%nNNs[i])
plt.plot(nPCs,acc,'--o',label='PCA')

plt.grid(axis='both')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

### 6) Implement ISOMAP, project the data set into 2D and color by label.
the idea is: performing MDS using as a distance an estimate of the geodetic distance in the embedded manifold

Initialize the estimate of the geodetic distance: delta_ij=d_ij if d_ij < d_c (i,j connected by a link), delta_ij=infty if d_ij > d_c (i,j not directly connected). In this way we infer the distance delta_ij on the embedded manifold from the one in the embedding manifold d_ij

```python
distances=euclidean_distances(x,x)
```

```python
# initialization of the distances through threshold value, but there are outliers...
dc=5
for i in range(size1):
    for j in range(i+1):
        if distances[i,j]>dc:
            distances[i,j]=np.inf
            distances[j,i]=np.inf
```

```python
# initialization of the distances through NNs
nNN=5 # after this, infinite distance
idx=distances.argsort(axis=1)
for i in range(size1):
    distances[i,:]=distances[i,idx[i]]
    distances[i,(nNN+1):]=np.inf
    idx2=idx[i,:].argsort()
    distances[i,:]=distances[i,idx2]

# symmetrize (due to infinities, i.e. link missing)
for i in range(size1):
    for j in range(size1):
        if(distances[i,j]<np.inf):
            distances[j,i]=distances[i,j]
```

```python
@njit(fastmath=True)
def geodetic(size1,distances):
    for k in range(size1):
        if (k%500==0):
            print(k)
        for i in range(size1):
            for j in range(size1):
                if distances[i,j]>distances[i,k]+distances[k,j]:
                    distances[i,j]=distances[i,k]+distances[k,j]
    return distances
```
