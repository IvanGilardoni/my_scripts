# Exercise 4: Markov state modeling
## The metastable states of a peptide in water solution
From the link https://github.com/alexdepremia/Unsupervised-Learning-Datasets.git you can download the file high variance distances.dat.gz. It corresponds to a time series of the value of distances between pair of atoms belonging to different residues in the peptide of sequence Arg-Phe-Phe-Glu-Ala. The chosen distances are all those that during the dynamics vary significantly (variance larger than 5  ̊A2. This peptide is a fragment of α-synuclein, and studying its conformational dynamics is important to understand Alzheimer disease.
Build a Markov State model from this time series

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from numba import njit
import os.path
import time
from scipy.spatial import distance_matrix
```

```python
data = np.loadtxt('high_variance_distances.txt', skiprows=0, unpack=True).T

npatt,nfeat=data.shape
print(data.shape)

means=np.zeros(nfeat)
stds=np.zeros(nfeat)

for i in range(nfeat):
    means[i]=np.mean(data[:,i])
    stds[i]=np.std(data[:,i])
    data[:,i]=(data[:,i]-means[i])/stds[i]
    plt.plot(data[:,i])
```

1. Find the microstates by a cluster analysis performed with k-means or k-medoids. Optionally, perform a preliminary dimensional reduction using PCA

```python
def draw(data,centroids,z,k,b,icount=0):
    data1=data[:,0]
    data2=data[:,1]
    
    plt.figure()
    for i in range(k):
        plt.plot(data1[np.where(z==i)],data2[np.where(z==i)],'o',label='%i'%i,markersize=2)
        if (b==0):
            plt.title('clustering initialization')
        else:
            plt.title('clustering at step %i'%icount)  
    plt.plot(centroids[:,0],centroids[:,1],'*k')
```

```python
# function for k-means
# bplot=0 if you do not want plots, 1 otherwise
njit(fastmath=True)
def kmeans(data,k,bplot=1):

    npatt,nfeat=data.shape
    
    # 1. randomly pick k centers
    centroids=np.zeros((k,nfeat))
    inds=np.random.choice(npatt,k,replace=False)
    for i in range(k):
        centroids[i,:]=data[inds[i],:]
        
    # 2. assign each point to its nearest center
    distances=distance_matrix(data,centroids)
    z=np.zeros(npatt) # column index of the min distance per row
    zold=np.zeros(npatt)
    z=np.argmin(distances,axis=1)      
    icount=0
    
    if (bplot==1):
        draw(data,centroids,z,k,b=0)
       
    while np.any(np.not_equal(z,zold)):
        zold=z
        # 3. update the k cluster centroids based on the assigned cluster membership
        for i in range(k):
            inds=np.where(z==i)
            centroids[i,:]=np.mean(data[inds,:],axis=1)
            
        distances=distance_matrix(data,centroids)
        z=np.argmin(distances,axis=1)
        icount+=1
        
        if (bplot==1):
            draw(data,centroids,z,k,b=1,icount=icount)
        
    obj=0
    for i in range(k):
        inds2=np.where(z==i)
        obj=obj+np.sum(np.linalg.norm(data[inds2,:]-centroids[i,:],axis=1)**2) 
            
    return z,centroids,obj
```

```python
cov=np.zeros((nfeat,nfeat))
cov=np.cov(data,rowvar=False)
```

```python
eigvals,eigvecs=np.linalg.eigh(cov)

idx=eigvals.argsort()[::-1]   
eigvals=eigvals[idx]
eigvecs=eigvecs[:,idx]

neigvals=eigvals/np.sum(eigvals) # "normalized" eigenvalues

plt.plot(neigvals,'*')
```

```python
plt.plot(neigvals,'*')
plt.xlim([-1,20])
plt.xticks(range(0,21))

xs=neigvals[0:14]
expl=np.sum(xs)
print(expl) # explained variance > 90% for 9 eigvals and > 95% for 14 eigvals 
del xs
```

```python
nPCs=14
dataPCs=np.zeros((npatt,nPCs))

for i in range(npatt):
    for j in range(nPCs):
        dataPCs[i,j]=np.sum(data[i,:]*eigvecs[:,j])
```

```python
f=plt.figure(figsize=(8,6))

plt.title("PCs projection", fontsize=20)

plt.xlabel("PC 1", fontsize=15)
plt.ylabel("PC 2", fontsize=15)

plt.plot(dataPCs[:,0],dataPCs[:,1],'*')
plt.show()

plt.figure()
plt.plot(dataPCs)
plt.title('PCs')
plt.show()
```

```python
N=5
kmin=2
kmax=30
nk=kmax-kmin+1

zs=np.zeros((nk,npatt))
objs=np.zeros(nk)

for k in range(kmin,kmax+1):
    objold=10**10
    for i in range(N):
        z,clusters,obj=kmeans(dataPCs,k,bplot=0)
        if (obj<objold):
            objold=obj
            [zs[k-kmin,:],objs[k-kmin]]=[z,obj]
```

```python
ks=np.arange(kmin,kmax+1)
plt.figure(figsize=(8,6))
plt.plot(ks,objs,'*-r',label='k-means')
plt.title('scree plot')
plt.xlabel('k')
plt.ylabel('obj function')
#plt.yscale('log')
plt.grid(axis='both')
plt.legend()
plt.show()
```

```python
ratios=np.zeros(nk-1)
for i in range(kmax-kmin):
    ratios[i]=objs[i+1]/objs[i]

plt.grid(axis='both')
plt.xlabel('n. of clusters k')
plt.ylabel('ratio obj(k)/obj(k+1)')
plt.plot(ks[0:-1],ratios,'*b')
plt.show()
```

```python
# the ratios' fit changes passing from 8 clusters to 9 clusters, so (guess) 8 clusters
kopt=4 # how many clusters

# take the best clustering with k=kopt clusters
z=zs[kopt-kmin,:]
centroids=np.zeros((kopt,nPCs))
for i in range(kopt):
    inds2=np.where(z==i)
    centroids[i,:]=np.mean(dataPCs[inds2,:],axis=1)

    
draw(dataPCs,centroids,z,kopt,b=0)
plt.title('best clustering')
plt.legend()
plt.show()
```

2. choose a time lag τ , and estimate the transition matrix between each pair of clusters π_(α,β) = P (β, τ |α, 0)

```python
tau=10 # how to choose the optimal value for tau?
ndtimes=npatt-tau # i.e. how many steps t --> t+tau
trans=np.zeros((kopt,kopt)) # the transition matrix
ntimecl=np.zeros(kopt) # i.e. #(alpha), n. of times in clusters alpha

w1=z[:-tau]
w2=z[tau:]

for i in range(kopt):
    ntimecl[i]=np.size(np.where(w1==i)) # w1 instead of z (only the ones with the corresponding exits t --> t+tau)
    for j in range(kopt):
        trans[i,j]=np.size(np.where((w1==i) & (w2==j))) # rows=entries, columns=exits
        trans[i,j]=trans[i,j]/ntimecl[i]

del w1,w2

print('time spent in each cluster:\n',ntimecl)
print('transition matrix:\n',trans)
print('total probs:\n',np.sum(trans,axis=1))

fig=plt.figure(figsize=(6,6))
matdraw=plt.matshow(trans,fignum=0,aspect='auto')
plt.title('transition matrix')
plt.xlabel('exits')
plt.ylabel('entries')
fig.colorbar(matdraw)
plt.show()
```

3. By analyzing the spectrum choose an appropriate number of Markov states. Find the Markov states by inspecting the sign of the eigenvectors.

```python
Teigvals,TReigvecs=np.linalg.eig(trans.T) # eigenvalues and RIGHT eigenvectors of the Transition matrix (T)

idx=(np.abs(Teigvals)).argsort()[::-1]   
Teigvals=Teigvals[idx]
TReigvecs=TReigvecs[:,idx]

print(Teigvals)
plt.plot(np.abs(Teigvals),'*')
plt.ylabel('|lambda|')
plt.grid(axis='both')
plt.show()
```

## CLAIM:
if an eigenvalue of a stochastic matrix is unique, then the corresponding eigenvector is real (i.e. it has only real components);

otherwise, if two eigenvalues are complex conjugated, then the corresponding eigenvectors are linear combinations (with real coefficients) of (i.e. they can be written as) two eigenvectors, one real and the other purely imaginary

Provided that, we can always identify Markov states

```python
print(TReigvecs.T)
#print(TReigvecs[:,1])
#print(TReigvecs[:,2])
ngamma=1 # number of time scales I consider

# for any cluster, give the corresponding Markov state
A=(np.sign(np.real(TReigvecs[0:kopt,1:ngamma+1]))+1)/2
markovst=np.zeros(kopt)
for i in range(ngamma):
    markovst=markovst+(2**i)*A[:,i]

print(A)
del A

print(markovst)
```

5. Discuss the stability of the results with respect to the metaparameters of the approach, in particular the time lag τ and the number of clusters.

```python
# given the chosen values for tau and kopt:
# compute the transition matrix, its eigenvalues and right eigenvectors;
# identify which clusters belong to the same Markov states, given the number of time scales I consider

def markov(npatt,tau,kopt,z,ngamma):
    
    ndtimes=npatt-tau # i.e. how many steps t --> t+tau
    trans=np.zeros((kopt,kopt)) # the transition matrix
    ntimecl=np.zeros(kopt) # i.e. #(alpha), n. of times in clusters alpha

    w1=z[:-tau]
    w2=z[tau:]

    for i in range(kopt):
        ntimecl[i]=np.size(np.where(w1==i)) # w1 instead of z (only the ones with the corresponding exits t --> t+tau)
        for j in range(kopt):
            trans[i,j]=np.size(np.where((w1==i) & (w2==j))) # rows=entries, columns=exits
            trans[i,j]=trans[i,j]/ntimecl[i]

    del w1,w2
    
    Teigvals,TReigvecs=np.linalg.eig(trans.T) # eigenvalues and RIGHT eigenvectors of the Transition matrix (T)

    idx=(np.abs(Teigvals)).argsort()[::-1]   
    Teigvals=Teigvals[idx]
    TReigvecs=TReigvecs[:,idx]
    
    for i in range(1,ngamma+1): # to have the same (?) identification of the Markov states
        if (np.real(TReigvecs[0,i])<0):
            TReigvecs[:,i]=-TReigvecs[:,i]
    
    A=(np.sign(np.real(TReigvecs[0:kopt,1:ngamma+1]))+1)/2
    markovst=np.zeros(kopt)

    for i in range(ngamma):
        markovst=markovst+(2**i)*A[:,i]

    del A
    
    return trans,Teigvals[1:ngamma+1],markovst
```

```python
#taus=[1,2,5,10,20,50,100,200,500]
taus=np.arange(1,3000,10)
leng=np.size(taus)
trans=np.zeros((leng,kopt,kopt))
Teigvals=np.zeros((leng,ngamma))
markovst=np.zeros((leng,kopt))

for i in range(leng):
    trans[i,:,:],Teigvals[i,:],markovst[i,:]=markov(npatt,taus[i],kopt,z,ngamma)

#plt.plot()
#print(Teigvals)
print(markovst)

plt.plot(taus,np.abs(Teigvals),'*-')
#plt.plot(taus,Teigvals,'*-')
```

```python
fig=plt.figure(figsize=(8,10))
matdraw=plt.matshow(markovst,fignum=0,aspect='auto',cmap='gray')
plt.title('Markov states')
plt.xlabel('cluster n.')
plt.ylabel('n. of iteration')
fig.colorbar(matdraw)
plt.show()
```
