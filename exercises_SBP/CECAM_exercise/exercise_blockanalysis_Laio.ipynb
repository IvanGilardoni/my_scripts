```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.loadtxt('ts3.dat')
```

```python
size=len(x)
print(size)
```

```python
#print(x)
plt.figure(figsize=(12,8))
plt.plot(x,'.')
#plt.plot(x[1500:1550],'.-')
plt.show()
mean=np.mean(x)
std=np.std(x)
print('mean=',mean)
print('std (on the mean)=',std/np.sqrt(size))
```

```python
#nbls=[2,4,5,6,7,8,9,10,12,15,16,18,20,22,25,26,28,30,32,34,36,38,40,44,48,50,60,70,80,90,100,150,200,300,400,500,750,1000,1500,2000,3000,4000,5000,7500,10000]#,20000]
#nbls=[2,5,10,15,20,30,40,50,60,70,80,90,100,150,200,300,400,500,750,1000,1500,2000,3000,4000,5000,7500,10000,20000]
#nbls=np.arange(10,10000,10)
#nbls=np.concatenate((np.arange(2,500,1),np.arange(550,10000,50)),axis=0) ## for ts1
#nbls=np.concatenate((np.arange(2,20,1),[40,50,70,90],np.arange(100,20000,100))) ## for ts2
nbls=np.arange(10,2000,10) ## for ts3

#nbls=np.arange(2,500,3)
epsilon=np.zeros(len(nbls))
for j in range(len(nbls)):
    sizebl=int(size/nbls[j])
    #print(sizebl)
    a=0
    for i in range(nbls[j]):
        a+=(np.mean(x[(sizebl*i):(sizebl*(i+1))]))**2
    epsilon[j]=np.sqrt((a/nbls[j]-mean**2)/nbls[j])
    #epsilon[sizebl-1]=a
plt.figure(figsize=(10,6))
plt.plot(nbls,epsilon,'.--')
plt.xlabel('number of blocks')
plt.xscale('log')
plt.grid()

print(np.mean(epsilon[:500]))
#print(epsilon)
```
