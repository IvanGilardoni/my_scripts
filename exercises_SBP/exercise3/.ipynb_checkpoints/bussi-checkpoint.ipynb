```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def MC(nsteps=10000,delta=0.1,q0=0.0):
    q=q0
    ene=0.5*q**2
    traj=[]
    for istep in range(nsteps):
        qtry=q+(2*np.random.rand()-1)*delta
        enetry=0.5*qtry**2
        acc=np.exp(-(enetry-ene))
        if acc>1.0:
            acc=1.0
        if acc>np.random.rand():
            q=qtry
            ene=enetry
        traj.append(q)
    return np.array(traj)
```

```python
np.std(MC())
```

```python
traj=MC()
plt.plot(traj)
```

```python
for delta in [0.01,0.1,1.0,10.0,100.0]:
    plt.plot(MC(delta=delta,nsteps=100),label=str(delta))
    
plt.legend()
plt.show()
```

```python
for q0 in [0.0,1.0,2.0,5.0,10.0,20.0,50.0]:
    plt.plot(MC(q0=q0,nsteps=1000,delta=0.1),label=str(q0))
    
plt.legend()
plt.show()
```

```python
def MC(nsteps=10000,delta=0.1,q0=0.0):
    q=q0
    ene=0.5*q**2
    traj=[]
    acceptance=[]
    acc1=0
    for istep in range(nsteps):
        qtry=q+(2*np.random.rand()-1)*delta
        enetry=0.5*qtry**2
        acc=np.exp(-(enetry-ene))
        if acc>1.0:
            acc=1.0
        if acc>np.random.rand():
            acc1+=1
            q=qtry
            ene=enetry
        traj.append(q)
        acceptance.append(acc)
    return np.array(traj),np.average(acceptance),acc1/nsteps
```

```python
acc=[]
acc1=[]
deltas=np.array([0.01,0.1,1.0,10.0,100.0])
for delta in deltas:
    acc.append(MC(delta=delta)[1])
    acc1.append(MC(delta=delta)[2])
```

```python
plt.plot(deltas,acc,"-x")
plt.plot(deltas,acc1,"-x")
plt.plot(deltas,np.exp(-0.4*deltas))

plt.xscale("log")
```

```python
def MC(nsteps=10000,delta=1.0,q0=None):
    if q0 is None:
        q0=np.zeros((2,3))
    
    q=+q0
    
    ene=0.5*(np.sqrt(np.sum((q[1]-q[0])**2))-4)**2
    traj=[]
    for istep in range(nsteps):
        i=np.random.randint(6)
        qtry=+q
        inc=(2*np.random.rand()-1)*delta
        if(i<3):
            qtry[0,i]+=inc
        else:
            qtry[1,i-3]+=inc

        enetry=0.5*(np.sqrt(np.sum((qtry[1]-qtry[0])**2))-4)**2
        acc=np.exp(-(enetry-ene))
        if acc>1.0:
            acc=1.0
        if acc>np.random.rand():
            q=+qtry
            ene=enetry
        traj.append(q)
    return np.array(traj)
```

```python
traj=MC()
```

```python
def write_xyz(traj):
    with open(file="traj.xyz",mode="w") as f:
        for i in range(traj.shape[0]):
            print(2,file=f)
            print("",file=f)
            print("Ar",traj[i,0,0],traj[i,0,1],traj[i,0,2],file=f)
            print("Ar",traj[i,1,0],traj[i,1,1],traj[i,1,2],file=f)
```

```python
write_xyz(traj)
```
